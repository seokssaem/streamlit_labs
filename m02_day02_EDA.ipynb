# 인코딩 (Encoding)
- 문자 데이터를 숫자로 바꾸는 과정

## Label Encoding (레이블 인코딩)
- 값에 숫자를 직접 매긴다.

```python
df
```

```python
df['gender_num'] = df['gender'].map({'F':0, 'M':1})
```

```python
df
```

## One-Hot Encoding (원-핫 인코딩)
- True / False
- 1 / 0

```python
df_encoded = pd.get_dummies(df, columns=['gender'])
```

```python
df_encoded
```

# 스케일링 
- 숫자 값의 범위를 조정
- ex. 나이 --> 20~70, 연봉 --> 30,000,000

## MinMaxScaler
- 사이킷런 안에 있다.
- 값을 0 ~ 1 범위로 바꾸어준다.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['height_scaled']] = scaler.fit_transform(df[['height']])
```

```python
df
```

## StandardScaler
- 평균이 0, 표준편차가 1이 되도록 조정
- 사이킷런 안에 있다.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['height_standard']] = scaler.fit_transform(df[['height']])
```

```python
df
```
